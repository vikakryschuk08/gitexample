```python
import matplotlib.pyplot as plt
import numpy as np

# Генеруємо дані для параболи
x = np.linspace(-5, 5, 100)  # Генеруємо 100 значень x від -5 до 5
y = x**2  # Обчислюємо відповідні значення y (y = x^2)

# Побудова параболи
plt.plot(x, y, label='Парабола: y = x^2')

# Додавання міток і легенди
plt.xlabel('x')
plt.ylabel('y')
plt.title('Парабола')
plt.legend()

# Відображення графіка
plt.grid(True)
plt.show()
```
